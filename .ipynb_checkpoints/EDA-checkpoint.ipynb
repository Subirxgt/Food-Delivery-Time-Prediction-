```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_dataset.csv')
```

```python
df.describe(), df.info()
```

```python
fig,axes = plt.subplots(1, 3, figsize=(20, 4))

#plot 1 :
sns.histplot(df['Delivery_person_Age'], bins=20, color='blue', edgecolor='black',kde=True,ax=axes[0])
axes[0].set_title('Delivery Person Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

#plot 2 :
sns.histplot(df['Delivery_person_Ratings'], bins=20, color='green', edgecolor='black',kde=True, ax=axes[1])
axes[1].set_title('Delivery Person Ratings Distribution')
axes[1].set_xlabel('Ratings')
axes[1].set_ylabel('Frequency')

#plot 3 :
# sns.countplot(data=df, x='multiple_deliveries', hue='multiple_deliveries', palette='viridis', legend=False, ax=axes[1, 0])
# axes[1, 0].set_title('Multiple Deliveries Count')
# axes[1, 0].set_xlabel('Multiple Deliveries')

#plot 4 :
sns.histplot(df['Time_taken(min)'], bins=20, color='orange', edgecolor='black', kde=True, ax=axes[ 2])
axes[ 2].set_title('Delivery Time taken distribution')
axes[ 2].set_xlabel('Time taken (min)')
axes[ 2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
features = ['Weatherconditions', 'Road_traffic_density', 'Type_of_order','Vehicle_condition',
            'Type_of_vehicle', 'Festival', 'City', 'multiple_deliveries']


fig, axes = plt.subplots(2,4, figsize=(20, 10))
axes = axes.flatten()

# Loop and plot
for i, col in enumerate(features):
    sns.countplot(x=col,hue=col, data=df, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Countplot of {col}')
    axes[i].tick_params(axis='x', rotation=45)

# Remove empty axes
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
features = ['Weatherconditions', 'Road_traffic_density', 'Type_of_order','Vehicle_condition',
            'Type_of_vehicle', 'Festival', 'City', 'multiple_deliveries']
fig, axes = plt.subplots(2,4, figsize=(20, 10))
axes = axes.flatten()
# Choose the target variable
target = 'Time_taken(min)'

# Loop and plot
for i, col in enumerate(features):
    sns.boxplot(x=col, y=target, data=df, ax=axes[i], hue=col,palette='magma')
    axes[i].set_title(f'{target} by {col}')
    axes[i].tick_params(axis='x', rotation=45)

# Remove empty axes
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
sns.catplot(x='Road_traffic_density', y='Time_taken(min)', hue='City', kind='box', data=df)
```

```python
sns.catplot(
    x='City',
    y='Time_taken(min)',
    hue='Festival',
    kind='box',
    data=df,
    height=5,
    aspect=1.5
)
```

```python
sns.lmplot(
    x='Delivery_person_Ratings',
    y='Time_taken(min)',
    data=df,
    height=6,
    aspect=1.3,
    line_kws={"color": "red"}
)
plt.title('Regression: Age vs Time Taken')
plt.show()
```

```python
# Higher-rated delivery persons are consistently quicker, regardless of age.
# Younger delivery persons tend to deliver faster, especially when highly rated.
# Worst performance: Young but low-rated

# Create bins
df['Age_bin'] = pd.cut(df['Delivery_person_Age'], bins=[15, 22, 26, 30, 35, 50])
df['Rating_bin'] = pd.cut(df['Delivery_person_Ratings'], bins=[2, 3, 3.5, 4, 4.5, 5])

# Pivot for heatmap
pivot = df.pivot_table(
    index='Age_bin',
    columns='Rating_bin',
    values='Time_taken(min)',
    aggfunc='mean'
)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Average Delivery Time by Age & Rating Bins')
plt.xlabel('Ratings')
plt.ylabel('Age Groups')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(
    x='Road_traffic_density',
    y='Time_taken(min)',
    hue='Festival',
    data=df,
    estimator='mean',
)

plt.title('Average Time by Traffic and Festival')
plt.ylabel('Mean Time Taken (min)')
plt.show()
```
